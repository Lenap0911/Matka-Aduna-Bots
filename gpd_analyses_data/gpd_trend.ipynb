```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import math
```

```python
# percentage change of gdp year to year
gdp_data=pd.read_csv("update_gdp.csv")
gdp_data["gdp_change"] = round(gdp_data["gdp"].pct_change() * 100, 2)
gdp_data.head()
```

```python
# Initialize the column with zeros
gdp_data["good_fishing_year"] = 0

# Set good fishing years starting from year 5 (index 4), then every 3 years
for i in range(0, len(gdp_data)):
    if i >= 4:  # Year 5 is at index 4
        # Calculate position relative to year 5
        years_since_first_good = i - 1
        # Every 3rd year starting from year 5 is a good fishing year
        if years_since_first_good % 3 == 0:
            gdp_data.loc[i, "good_fishing_year"] = 1

gdp_data.head(10)


```

```python
# Initialize the column with zeros
gdp_data["ex_good_farming_year"] = 0

# Year 100 is the last of 6 good years, so years 95-100 are good
# Pattern: 1 bad year, then 6 good years (repeating)
# Working backwards from year 100 (index ): years 95-100 are good
# Years 88-94: 88 bad, 89-94 good
# Years 81-87: 81 bad, 82-87 good
for i in range(0, len(gdp_data)):
    # Year 100 is at index 99
    # Calculate position in the 7-year cycle (1 bad + 6 good)
    # Year 100 (index 99) should be position 6 in its cycle
    position_in_cycle = (i - 99 + 5) % 7
    
    # If position is 1-6 (not 0), it's a good farming year
    if position_in_cycle != 0:
        gdp_data.loc[i, "ex_good_farming_year"] = 1

gdp_data.tail(50)
```

```python
#is the year actually a good farming year 
predictor= pd.read_csv("../data/data/predictors.csv")
gdp_data = gdp_data.merge(predictor[['year','farm_income']], on='year', how='left')


# Create a new column for actual good farming years
gdp_data["actual_good_farming_year"] = 0

# If it is predicted to be a good farming year AND the income is greater than or equal to average, then it is actually a good year
threshold= gdp_data["farm_income"].mean()-gdp_data["farm_income"].std()
actual_good_condition = (gdp_data["ex_good_farming_year"] == 1) & (gdp_data["farm_income"] >= threshold)
gdp_data.loc[actual_good_condition, "actual_good_farming_year"] = 1
gdp_info = gdp_data[["year","gdp","gdp_change","good_fishing_year","ex_good_farming_year","actual_good_farming_year"]]
gdp_data.tail(30)
```

```python
#now i want to know what is the averege percentage difference between a farmers income for a good and bad year compared to the previous year


# Calculate percentage change in farm income from previous year
gdp_data["farm_income_change"] = gdp_data["farm_income"].pct_change() * 100

# Separate good and bad years
good_years = gdp_data[gdp_data["actual_good_farming_year"] == 1]
bad_years = gdp_data[gdp_data["actual_good_farming_year"] == 0]

# Average percentage change for good years (compared to previous year)
avg_change_good = good_years["farm_income_change"].mean()

# Average percentage change for bad years (compared to previous year)
avg_change_bad = bad_years["farm_income_change"].mean()

print(f"Average % change in farm income for good years: {avg_change_good:.2f}%")
print(f"Average % change in farm income for bad years: {avg_change_bad:.2f}%")
print(f"Difference: {avg_change_good - avg_change_bad:.2f}%")

```

```python
# Get farm income change for good years that follow bad years
gdp_data["prev_year_type"] = gdp_data["actual_good_farming_year"].shift(1)

# Good years that follow bad years (prev year = 0, current year = 1)
bad_to_good = gdp_data[(gdp_data["actual_good_farming_year"] == 1) & (gdp_data["prev_year_type"] == 0)]

# Average change when transitioning from bad to good year
avg_bad_to_good = bad_to_good["farm_income_change"].mean()

print(f"Average % increase when going from bad to good year: {avg_bad_to_good:.2f}%")
```

```python
# Bad years that follow good years (prev year = 1, current year = 0)
good_to_bad = gdp_data[(gdp_data["actual_good_farming_year"] == 0) & (gdp_data["prev_year_type"] == 1)]

# Average change when transitioning from good to bad year
avg_good_to_bad = good_to_bad["farm_income_change"].mean()

print(f"Average % change when going from good to bad year: {avg_good_to_bad:.2f}%")
```
